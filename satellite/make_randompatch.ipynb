```python
#Tanmay urf Gilfoyle method of giving random path to image
def RandomPatch(xld, xhd, yld, yhd):
    a = randint(0,256)
    b = randint(0,256)
    c = randint(0,256)
    d = randint(0,256)
    diff1 = b-a
    diff2 = d-c
    
    while(not(xld<diff1<xhd) or not(yld<diff2<yhd)):
        a = randint(0,256)
        b = randint(0,256)
        c = randint(0,256)
        d = randint(0,256)
        diff1 = b-a
        diff2 = d-c        
        
    return a,b,c,d
```
